# Content Based Filtering

```python
# dfm['genres'][1].split("|")

# 2. count vectorizer: hitung kata tiap data

cv = CountVectorizer()
cm = cv.fit_transform(dfm['genres'])

# total kata unik 
print(cv.get_feature_names())

# matrix freq tiap kata di tiap data
print(cm.toarray())
# print(cm.toarray()[0])
```

```python
# 3. Cosine similarity dari tiap data di matrix freq

cosScore = cosine_similarity(cm.toarray())
cosScore
```

```python
# 4. Gunakan cos score sbg data rekomendasi
sayaSuka = dfm[dfm['title'] == 'Superman vs. The Elite (2012)'].index.values[0]

# daftar seluruh Movie beserta cos score
similarMovie = list(enumerate(cosScore[sayaSuka]))
# similarCars  # (index, %similarity)
```

```python
# sort berdasarkan %similarity
similarMovie = sorted(similarMovie, key=lambda x: x[1], reverse=True)
# similarCars
```

```python
# 5 Movie yang mirip


dfSim = []
for i in similarMovie[:5]:
    dfSim.append(dfm.iloc[i[0]])
dfSim = pd.DataFrame(dfSim)
dfSim
```
